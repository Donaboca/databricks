# Noudetaan rautatieasemien nimet dataframeen

```python
import pandas as pd
import requests as req
```

```python
url = 'https://rata.digitraffic.fi/api/v1/metadata/stations'

response = req.get(url)
data = response.json()
df = pd.DataFrame(data)
df = df[['stationName', 'stationShortCode']]
df.head()
```

## Tallennus .csv-muotoon

```python
df.to_csv('VR_traffic_points.csv', index=False)
```
